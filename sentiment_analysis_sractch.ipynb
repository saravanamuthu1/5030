```python
logprior = np.log(positive_frequency/negative_frequency)
print(logprior)
```

```python
freq_pos = Counter(" ".join(df.loc[df.label=='1']['tweet'].values.tolist()).split(" ")) 
print(freq_pos)
freq_neg = Counter(" ".join(df.loc[df.label=='0']['tweet'].values.tolist()).split(" ")) 
print(freq_neg)
print(type(len(freq_neg)))
```

```python
def pos_word_loglikelihood(w):
    if w in freq_pos:
        pos_total_probaility = (len(freq_pos)/ freq_pos.get(w,0)+1)
        p_w_pos = (pos_total_probaility* freq_pos.get(w,0))/len(freq_pos)
        return p_w_pos     
    else:
        return(0)
```

```python
def neg_word_loglikelihood(w):
  if w in freq_neg:
        neg_total_probaility = (len(freq_neg)/ freq_neg.get(w,0)+1)
        p_w_neg = (neg_total_probaility* freq_neg.get(w,0))/len(freq_neg)
        return p_w_neg      
  else:
        return(0)
```

```python
print(pos_word_loglikelihood("hahaha"))
print(neg_word_loglikelihood("hahaha"))
```

```python
def pos_text_loglikelihood(mytxt):
    pos_score = logprior
    for w in mytxt:
        pos_score*= pos_word_loglikelihood(w)
    return pos_score
```

```python
def neg_text_loglikelihood(mytxt):

    neg_score = logprior
    for w in mytxt:
        neg_score*= pos_word_loglikelihood(w)
    return neg_score
```

```python
df['score1'] = df.tweet.apply(lambda x:pos_text_loglikelihood(x))
df['score2'] = df.tweet.apply(lambda x:neg_text_loglikelihood(x))
df["prediction"] = df[["score1", "score2"]].max(axis=1) 



df['prediction'] = df.apply(lambda x : 0 if x['prediction'] in df['score1'] else 1, axis=1)
```

```python
df['prediction'].value_counts()
```

```python
df['label'] = df.label.apply(lambda x:int(x))
```

```python
accuracy = accuracy_score(df['label'], df['prediction'])
```
